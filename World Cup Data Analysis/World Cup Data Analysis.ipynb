```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#source of dataset: https://www.kaggle.com/abecklas/fifa-world-cup
wc =pd.read_csv('WorldCupMatches.csv')
wc
```

```python
#Data Cleaning (Removing missing data)
wc.dropna()
```

```python
#change the starting index from 0 to 1
wc.index = wc.index + 1
```

```python
#List of columns
wc.columns
```

```python
#deleting the columns that I don't need (for easier read)
del wc['Referee']
del wc['Assistant 1']
del wc['Assistant 2']
del wc['RoundID']
del wc['MatchID']
```

```python
#List of columns
wc.columns
```

```python
#Adding a column to specify the match
wc['Match Between'] = wc['Home Team Name']+ ' - ' +wc['Away Team Name']
wc
```

```python
#adding a column to display winner of the match
class WinningTeam:
    def __init__(self,row):
        self.row=row
    def winner(row):
        if row['Home Team Goals'] > row['Away Team Goals']:
            return  row['Home Team Name']
        elif row['Away Team Goals'] > row['Home Team Goals']:
            return row['Away Team Name']
        else:
            return 'Draw'
```

```python
wc['Winner of The Game'] = wc.apply(WinningTeam.winner, axis=1)
wc
```

```python
#Adding a column to indicate the goal difference in a game
wc['Goal Difference'] = ((wc['Home Team Goals']-wc['Away Team Goals']).abs())
wc
```

```python
#which city has hosted a world cup match the most?
wc['City'].value_counts().head(15)
```

```python
#which stadium has hosted a world cup match the most?
wc['Stadium'].value_counts().head(15)
```

```python
#maximum number of goals by a team in a single game:
wc[['Home Team Goals' , 'Away Team Goals']].max()
```

```python
#high
```

```python
top10_home_goals = wc.nlargest(10, 'Home Team Goals')
#top10_home_goals
```

```python
top10_away_goals = wc.nlargest(10, 'Away Team Goals')
#top10_away_goals
```

```python
top_10_goals = pd.concat([top10_home_goals, top10_away_goals])
top_10_goals[['Home Team Name','Away Team Name', 'Home Team Goals', 'Away Team Goals', 'Year']]
```

```python
#list matches based on the highest number of attendance
wc.sort_values('Attendance', ascending=False)[['Match Between', 'Winner of The Game', 'Attendance', 'Stadium', 'City' ,'Year']].head(30)
```

```python
#Sum of goals scored in each game
wc['sum_goals'] = wc['Home Team Goals'] + wc['Away Team Goals']
```

```python
#highest goals scored in a game
wc.sort_values('sum_goals', ascending=False)[['sum_goals','Match Between', 'Winner of The Game', 'Home Team Goals', 'Away Team Goals','Attendance', 'Year']].head(10)
```

```python
#which 15 teams have won the most games?
wc['Winner of The Game'].value_counts().head(15)
```

```python
#Sorting based on the goal diffrence
wc.sort_values(['Goal Difference'], ascending=False)[['Goal Difference','Match Between', 'Home Team Goals', 'Away Team Goals', 'Winner of The Game', 'Year']].head(30)
```

```python
#Years with most goals


goals_each_year = (wc.groupby('Year')['sum_goals'].sum().sort_values(ascending=False))
```

```python
goals_each_year
```

```python
# Displaying the years with most goals in a barchart
goals_each_year.plot.bar(figsize=(20,10))
```

```python
#display the top 10 teams with the highest number of wins
wc.drop(wc.loc[wc['Winner of The Game']=='Draw'].index, inplace=True)
wc['Winner of The Game'].value_counts().head(10).plot.pie(figsize=(20,10))
```
